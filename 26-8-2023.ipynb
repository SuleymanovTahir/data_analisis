```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_to_path='C:/Users/samsung/Desktop/Тахир папка/данные/uber/Players.csv'
file_to_path2='C:/Users/samsung/Desktop/Тахир папка/данные/uber/Teams.csv'
```

```python
teams=[]
players=[]
with open(file_to_path2) as f:
    rows=csv.DictReader(f)
    for r in rows:
        teams.append(r)
with open(file_to_path,encoding='utf-8') as f:
    rows=csv.DictReader(f)
    for r in rows:
        players.append(r)
```

```python
colors=[]
x=[]
y=[]
for player in players:
    x.append(float(player['minutes']))
    y.append(float(player['passes']))
    if (float(player['passes'])/float(player['minutes']))*100<50:
        colors.append('blue')
    elif (float(player['passes'])/float(player['minutes']))*100>=50 and (float(player['passes'])/float(player['minutes']))*100<70:
        colors.append('red')
    else:
        colors.append('yellow')

plt.scatter(x,y,c=colors)
plt.xlabel('Minutes')
plt.ylabel('Passes')
```

```python
bars = ['zero', 'one', 'two'] # number of red cards
heights = [0,0,0]
for red in teams:
    if int(red['redCards'])==0:
        heights[0]+=1
    elif int(red['redCards'])==1:
        heights[1]+=1
    elif int(red['redCards'])>=2:
        heights[2]+=1
plt.xticks([0,1,2],bars)
plt.bar([0,1,2],heights)
```

```python
explode=(0.1,0.1,0.1)
colors=['red','green','blue']
textprops={'color':'white','weight':'bold'}
plt.pie(heights,labels=bars,autopct='%1.1f%%',explode=explode,colors=colors,textprops=textprops,startangle=90,shadow=True)
```
